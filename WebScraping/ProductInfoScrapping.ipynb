```python
# Function to retrieve the mapping values 
def get_product_info(row):
    product_name = row['Product Name']
    brand = row['Brand']  # Assuming 'Brand' column exists
    
    # If the product is from Skindeep, update only Skin Type and Benefit from mapping
    if brand == 'Skindeepintl':
        info = mapping.get(product_name, {})
        return {
            'Category': row['Category'],  # Keep existing Category
            'Skin Type': info.get('Skin Type', row.get('Skin Type', None)),  # Update Skin Type from mapping if available
            'Benefit': info.get('Benefit', row.get('Benefit', None)),  # Update Benefit from mapping if available
            'Ingredients': row['Ingredients'],  # Keep existing Ingredients
        }
    
    # For non-Skindeep products, pull all values from the mapping
    info = mapping.get(product_name, {})
    return {
        'Category': info.get('Category', row.get('Category', None)),  # Use existing if not found in mapping
        'Skin Type': info.get('Skin Type', row.get('Skin Type', None)),
        'Benefit': info.get('Benefit', row.get('Benefit', None)),
        'Ingredients': info.get('Ingredients', row.get('Ingredients', None)),
    }

# Apply the function to each row and extract the details
df[['Category', 'Skin Type', 'Benefit', 'Ingredients']] = df.apply(get_product_info, axis=1, result_type='expand')

# Print the updated DataFrame
print(df)
```

```python
# Define the conversion rate for USD to PKR
usd_to_pkr_rate = 280  # Adjust this as per the current exchange rate

# Function to extract the numeric price value and currency
def extract_price_and_currency(price):
    price = str(price).strip()  # Ensure it's a string and remove extra spaces
    if '$' in price:
        # Extract numeric value for USD prices
        price_cleaned = ''.join(c for c in price if c.isdigit() or c == '.')
        try:
            return float(price_cleaned), 'USD'
        except ValueError:
            return None, None
    elif 'Rs' in price or 'PKR' in price:
        # Extract numeric value for PKR prices, removing commas and dots used as thousand separators
        price_cleaned = ''.join(c for c in price if c.isdigit() or c == '.').replace(',', '').replace('.', '')
        try:
            return float(price_cleaned), 'PKR'
        except ValueError:
            return None, None
    return None, None

# Function to convert USD to PKR (keeping Skindeep prices unchanged)
def convert_usd_to_pkr(row):
    if row['Brand'] == 'Skindeepintl':  # Keep price unchanged for Skindeep products
        return row['Price']
    if row['currency'] == 'USD':  # If the price is in USD
        return row['Price_Value'] * usd_to_pkr_rate  # Convert to PKR
    return row['Price_Value']  # If the price is already in PKR, keep it

# Function to apply extraction and conversion for price
def process_price(row):
    # Extract price and currency for the row
    price_value, currency = extract_price_and_currency(row['Price'])
    row['Price_Value'] = price_value
    row['currency'] = currency
    row['price_in_pkr'] = convert_usd_to_pkr(row)
    return row

# Apply the price processing function to each row
df = df.apply(process_price, axis=1)

# Verify the results
print(df[['Price', 'Price_Value', 'currency', 'price_in_pkr']])
```

```python
#Standardizing product names
from rapidfuzz import fuzz, process
# Define a function to standardize product names
def standardize_product_names(product_names, threshold=85):
    standardized_names = {}
    unique_names = list(set(product_names))
    
    for name in unique_names:
        # Find the closest match in the standardized names
        match = process.extractOne(name, standardized_names.keys(), scorer=fuzz.token_sort_ratio)
        
        if match and match[1] >= threshold:  # If a close match is found
            standardized_names[name] = match[0]
        else:  # If no close match, use the current name as the standard
            standardized_names[name] = name

    # Map all product names to their standardized name
    return [standardized_names[name] for name in product_names]

# Apply standardization
Product_information['Standardized Product Name'] = standardize_product_names(Product_information['Product Name'])
# Drop duplicates based on standardized product names
df_deduplicated = Product_information.drop_duplicates(subset='Standardized Product Name').reset_index(drop=True)



# Output the deduplicated DataFrame
df_deduplicated.head()
df_deduplicated.info()
```
